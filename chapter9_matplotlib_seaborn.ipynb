```python
def prime(n):
    bool_list=[True]*(n+1)    
    res=[]
    for i in range(2,n):
        if bool_list[i]:
            res+=[i]     #将该质数加入
            for j in range(i*2,n+1,i):  #此步是在i*2到n+1的范围内，每隔i个
                bool_list[j]=False  #把刚才加入的i，其它i的倍数定为False，    
                #例子:i=3,
                #for j in range(3*2,n+1,3):
                #即6，9，12...这些去除的bool_list[j]=False
    return res
```

```python
print(prime(50))
```
